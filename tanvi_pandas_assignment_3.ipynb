### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users= df = pd.read_csv("u.user.txt", sep='|')
users.head(5)
```

### Step 4. Discover what is the mean age per occupation

```python
users.head(5)
occ= df.groupby(['occupation'])
age= occ['age'].mean()
print(age)
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
gender= occ['gender'].value_counts().unstack()
male_ratio= gender['M']/gender['F']
male_ratio= male_ratio.sort_values(ascending=False)
print(male_ratio)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
max_age= occ['age'].max()
min_age= occ['age'].min()
print(max_age, min_age)
```

### Step 7. For each combination of occupation and gender, calculate the mean age. Also print this without hierarchical indexing

```python
group= users.groupby(['occupation', 'gender'])
mean_age= group['age'].mean().unstack()
print(mean_age)

```
